```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import statsmodels.tsa.stattools as ts 
import csv
```

```python
cat_f = open('../data/day/CAT-1d.csv', 'r')
cmi_f = open('../data/day/CMI-1d.csv', 'r')
spy_f = open('../data/day/SPY-1d.csv', 'r')
pg_f = open('../data/day/PG-1d.csv', 'r')
aapl_f = open('../data/day/AAPL-1d.csv', 'r')
db_f = open('../data/day/DB-1d.csv', 'r')
cat = np.array(list(csv.reader(cat_f)))[:, 4].astype(np.float64)
cmi = np.array(list(csv.reader(cmi_f)))[:, 4].astype(np.float64)
spy = np.array(list(csv.reader(spy_f)))[:, 4].astype(np.float64)
pg = np.array(list(csv.reader(pg_f)))[:, 4].astype(np.float64)
aapl = np.array(list(csv.reader(aapl_f)))[:, 4].astype(np.float64)
db = np.array(list(csv.reader(db_f)))[:, 4].astype(np.float64)
```

```python
plt.plot(cat)
plt.plot(cmi)
plt.plot(spy)
plt.plot(pg)
plt.plot(aapl)
plt.plot(db)
plt.show()
```

```python
def price_to_ret(price):
    ret_data = []
    for i in range(1, price.shape[0]):
        ret = (price[i] - price[i - 1]) / price[i - 1]
        ret_data.append(ret)
    return np.array(ret_data)
```

```python
cat_ret = price_to_ret(cat)
cmi_ret = price_to_ret(cmi)
spy_ret = price_to_ret(spy)
plt.plot(cat_ret)
plt.plot(cmi_ret)
plt.plot(spy_ret)
plt.show()
```

```python
result = ts.coint(cmi, cat)
print(result)
```

```python
diff = cmi - cat
plt.plot(diff)
plt.show()
```
