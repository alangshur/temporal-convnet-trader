```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import csv
```

```python
cat_f = open('../data/day/CAT-1d.csv', 'r')
cmi_f = open('../data/day/CMI-1d.csv', 'r')
spy_f = open('../data/day/SPY-1d.csv', 'r')
cat = np.array(list(csv.reader(cat_f)))[:, 4].astype(np.float64)
cmi = np.array(list(csv.reader(cmi_f)))[:, 4].astype(np.float64)
spy = np.array(list(csv.reader(spy_f)))[:, 4].astype(np.float64)
```

```python
print('CAT length: {}'.format(cat.shape[0]))
print('CMI length: {}'.format(cmi.shape[0]))
print('SPY length: {}'.format(spy.shape[0]))
```

```python
plt.plot(cat)
plt.plot(cmi)
plt.plot(spy)
plt.show()
```

```python
ratio = cmi / (cat * spy)
plt.plot(ratio)
plt.show()
```

```python
def price_to_ret(price):
    ret_data = []
    for i in range(1, price.shape[0]):
        ret = (price[i] - price[i - 1]) / price[i - 1]
        ret_data.append(ret)
    return np.array(ret_data)
```

```python
cat_ret = price_to_ret(cat)
cmi_ret = price_to_ret(cmi)
spy_ret = price_to_ret(spy)
plt.plot(cat_ret)
plt.plot(cmi_ret)
plt.plot(spy_ret)
plt.show()
```

```python
ratio_ret = price_to_ret(ratio)
plt.plot(ratio_ret)
plt.show()
```

```python
print('Correlation between CMI and CAT: {}'.format(np.corrcoef(cmi_ret, cat_ret)))
```

```python
plot_acf(ratio_ret)
plt.show()
```
