```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import statsmodels.tsa.stattools as ts 
import csv
```

```python
spy_f = open('data/day/SPY-1d.csv', 'r')
aapl_f = open('data/day/AAPL-1d.csv', 'r')
gld_f = open('data/day/GLD-1d.csv', 'r')
spy = np.array(list(csv.reader(spy_f)))[:, 4].astype(np.float64)
aapl = np.array(list(csv.reader(aapl_f)))[:, 4].astype(np.float64)
gld = np.array(list(csv.reader(gld_f)))[:, 4].astype(np.float64)
```

```python
plt.plot(spy)
plt.plot(aapl)
plt.plot(gld)
plt.show()
```

```python
def price_to_ret(price):
    ret_data = []
    for i in range(1, price.shape[0]):
        ret = (price[i] - price[i - 1]) / price[i - 1]
        ret_data.append(ret)
    return np.array(ret_data)
```

```python
spy_ret = price_to_ret(spy)
aapl_ret = price_to_ret(aapl)
gld_ret = price_to_ret(gld)
plt.plot(spy_ret)
plt.plot(aapl_ret)
plt.plot(gld_ret)
plt.show()
```

```python
period = 30
stds = []

for i in range(period, spy_ret.shape[0]):
    std = spy_ret[i - period:i].std()
    stds.append(std)
    
plt.plot(stds)
plt.show()
```
